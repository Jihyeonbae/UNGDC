- The goal is to process texts in a unit smaller than the entire document which has to be truncated to meet the token limit.

```python
import pandas as pd

file_path = '../../../data/raw/txt_files/JPN_45_1990.txt'

# Open the file in read mode
with open(file_path, 'r', encoding='utf-8') as file:
    # Read the content of the file
    text = file.read()

# Create a DataFrame with a single column named 'text'
df = pd.DataFrame({'text': [text]})

# Display the DataFrame
print(df)
```

```python
tt = TextTilingTokenizer(w=20, k=10,similarity_method=0, stopwords=None, smoothing_method=[0], smoothing_width=2, smoothing_rounds=1, cutoff_policy=1, demo_mode=False)

# Tokenize the text using TextTiling
tt = TextTilingTokenizer(demo_mode=True)
tiles, ss, d, b = tt.tokenize(text)

# Display paragraph boundaries
print("Paragraph Boundaries:", b)

```

```python
# Iterate through the paragraph indices and print paragraphs
for i in range(len(b) - 1):
    paragraph_start = b[i]
    paragraph_end = b[i + 1]
    paragraph = text[paragraph_start:paragraph_end].strip()
    print(f"Paragraph {i + 1}:\n{paragraph}\n")
```

```python
# Extract paragraphs based on boundaries
paragraphs = [text[b[i-1]:b[i]] for i in range(1, len(b))]

# Display paragraphs
for i, paragraph in enumerate(paragraphs, start=1):
    print(f"Paragraph {i}:\n{paragraph}\n")
```
